```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
# Load dataset
df = pd.read_csv("../data/mental_health_prediction.csv")

# Display first rows
print("First 5 rows:")
print(df.head())

# Convert Gender to numeric for correlation
df["Gender"] = df["Gender"].map({
    "Male": 0,
    "Female": 1
})

# -----------------------------
# Depression Distribution
# -----------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="Depression", data=df)
plt.title("Depression Distribution")
plt.show()

# -----------------------------
# Sleep vs Depression
# -----------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x="Depression", y="Sleep_Hours", data=df)
plt.title("Sleep Hours vs Depression")
plt.show()

# -----------------------------
# Stress vs Depression
# -----------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x="Depression", y="Stress_Level", data=df)
plt.title("Stress Level vs Depression")
plt.show()

# -----------------------------
# Correlation Heatmap
# -----------------------------
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation")
plt.show()
```
